### Ejercicio 1: 
Seguir un procedimiento similar al mostrado cuando se generó una distribución binomial, pero en esta caso genere un código que genere variables aletorias Poisson cuya función de distribución de probabilidad esta dada por:
>$$P(k,\lambda)=\frac{e^{-\lambda}(\lambda)^k}{k!}$$
    
> Demuestre matemáticamente que 
> $$P(k+1)=\frac{\lambda}{k+1}P(k)$$
> y a partir de esta relación genere variables aletorias que distribuyen poisson usando el método de la transformada inversa.

### Código de solución 
$$P(k,\lambda)=\frac{e^{-\lambda}(\lambda)^k}{k!}$$
$$P(k + 1)=\frac{e^{-\lambda}(\lambda)^{k+1}}{(k+1)!}$$
$$P(k + 1)=\frac{e^{-\lambda}(\lambda)^{k}\lambda}{(1+k)k!}$$
$$P(k + 1)=\frac{\lambda}{(1+k)} P(k,\lambda) $$

### Respuesta planteada en el ejercicio usando
 La respuesta es $$P(k + 1)=\frac{\lambda}{(1+k)} P(k,\lambda) $$

```python
def Poisson(lambd, k):
    return (np.exp((-lambd)) * lambd**k / sps.gamma(k))

lambd = 10
k = np.arange(0,30)
poisson_dat = Poisson(lambd, k)
plt.hist(poisson_dat, bins=10, density=True, color="orange")
plt.xlabel('Valores aleatorios')
plt.ylabel('Probabilidad')
plt.title('Distribución con Poisson')
plt.show()
```

```python
# Código de solución 
def Acep_rech2(R2:'Variables distruidas U~U(0,1)',
                 R1:'Variables distribuidas como g(x)',
                 f:'función objetivo a generar',
                 t:'función que mayora a f'):
#   R1 = np.random.rand(N)
    f_x = f(R1)
    t_x = t(R1)
    w = np.zeros(len(R1))
    condition = np.multiply(R2,t_x) <= f_x
    w = np.asarray([1 if condition[i] else 0 for i in range(len(R1))]).sum()
    return w
```

```python
aceptar_senoidal = Acep_rech2(R2,R1,f,t) / len(R1)
print('El porcentaje de puntos aceptados cuando usamos t(x) senoidal es: ', aceptar_senoidal * 100, '%') 
```
